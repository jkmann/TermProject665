```python
PATH = os.getcwd()
data_path = PATH + '/DatasetOfRacesMixed'
data_dir_list = os.listdir(data_path)
data_dir_list.remove('.DS_Store')
print (data_dir_list)
```

```python
img_data_list=[]
img_path_list=[]
```

```python
for dataset in data_dir_list:
    img_list=os.listdir(data_path+'/'+ dataset)
    print ('Loaded the images of dataset-'+'{}\n'.format(dataset))
    for img in img_list:
        if img is '.DS_Store':
            None
        else:
            img_path = data_path + '/'+ dataset + '/'+ img
            img_path_list.append(img_path)
            img = image.load_img(img_path, target_size=(224, 224))
            x = image.img_to_array(img)
            x = np.expand_dims(x, axis=0)
            x = preprocess_input(x)
            print('Input image shape:', x.shape)
            img_data_list.append(x)
    print (len(img_data_list))
```

```python
img_data = np.array(img_data_list)
print (img_data.shape)
img_data=np.rollaxis(img_data,1,0)
print (img_data.shape)
img_data=img_data[0]
print (img_data.shape)
```

```python
num_classes = 8
num_of_samples = img_data.shape[0]
labels = np.ones((num_of_samples,),dtype='int64')
```

```python
print(img_path_list[0])
print(img_path_list[80])
print(img_path_list[160])
print(img_path_list[240])
print(img_path_list[320])
print(img_path_list[400])
print(img_path_list[480])
print(img_path_list[560])
```

```python
labels[0:80]=0
labels[80:160]=1
labels[160:240]=2
labels[240:320]=3
labels[320:400]=4
labels[400:480]=5
labels[480:560]=6
labels[560:]=7
```

```python
names = ['HispanicFemale','CaucasianFemale','AfricanAmericanFemale','CaucasianMale', 'AfricanAmericanMale', 'AsianMale', 'AsianFemale', 'HispanicMale']
```

```python
Y = np_utils.to_categorical(labels, num_classes)
```

```python
x,y = shuffle(img_data,Y, random_state=2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
classifier = svm.SVC(kernel='linear', C=0.01)
y_pred = classifier.fit(X_train, y_train).predict(X_test)
```
